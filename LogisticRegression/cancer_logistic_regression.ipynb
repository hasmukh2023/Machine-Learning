```python
def predict(data,coef):
  yhat=coef[0]
  high=len(data.columns.tolist())
  for i in range(0,len(data)):
    for j in range(0,high):
      yhat+=coef[j+1] * data.iat[i,j]  
    # print(yhat)   
  return 1/(1+math.exp(-yhat))
    
  
```
