```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import  KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score ,confusion_matrix
```

```python
df=pd.read_csv('./titanic.csv')
df
```

```python
df.columns
```

```python
df.shape
```

```python
df1=df.drop(columns=['cabin','body','home.dest','boat'],axis=1)
df1
```

```python
unique_val=[]
for cols in df1.columns:
    unique_val.append(len(df1[cols].value_counts()))
unique_val
unique_val_df=pd.DataFrame({'column':df1.columns.tolist(),'unique_val':unique_val})
unique_val_df
```

```python
df1.isnull().sum()
```

```python
df1=df1.drop(1309)
df1
```

```python
df1['fare']=df1['fare'].fillna(df1['fare'].mean())
df1.isnull().sum()
```

```python
df1['embarked'].value_counts()
```

```python
df1['embarked']=df1['embarked'].fillna(df1['embarked'].mode()[0])
df1.isnull().sum()                                       
```

```python
location=df1['embarked'].map({'S':0,'C':1,'Q':2})
location
```

```python
df1.insert(2,'Location',location)
```

```python
pclass1=df1[df1['pclass']==1]
pclass1.head(5)
```

```python
pclass2=df1[df1['pclass']==2]
pclass2.head(5)
```

```python
pclass3=df1[df1['pclass']==3]
pclass3.head(5)
```

```python
pclass1['age']=pclass1['age'].fillna(pclass1['age'].mean())

pclass2['age']=pclass2['age'].fillna(pclass2['age'].mean())
pclass3['age']=pclass3['age'].fillna(pclass3['age'].mean())
```

```python
df2=pd.concat([pclass1,pclass2,pclass3])
```

```python
df2.isnull().sum()
```

```python
outlier_cols=df2[['age']]
```

```python
outliers=pd.DataFrame(columns=['columns','outliers'])
outliers
```

```python
for cols in outlier_cols:
    q1=df2[cols].quantile(0.25)
    q3=df2[cols].quantile(0.75)
    iqr=q3-q1
    low=q1 - iqr * 1.5
    high=q3 + iqr*1.5
    outliers=outliers.append({'columns':cols,'outliers':df2.loc[(df2[cols] > high) | (df2[cols]<low)].shape[0]},ignore_index=True)
outliers    
```

```python
df2


```

```python
x=df2[['pclass','age','sibsp','parch','fare','Location']]
x
```

```python
df2['ticket'].value_counts()
```

```python
y=df2[['survived']]
y=y.astype(int)
y
```

```python
x_train1,x_test1,y_train1,y_test1=train_test_split(x,y,test_size=0.2,random_state=1)

print(x_train1.shape,x_test1.shape)
print(y_train1.shape,y_test1.shape)

```

```python
np.sqrt(len(x_test1))

```

```python
model1=KNeighborsClassifier(n_neighbors=16,p=2,metric='euclidean')
```

```python
model1.fit(x_train1,y_train1)
```

```python
y_pre1=model1.predict(x_test1)
```

```python
acc1=accuracy_score(y_test1,y_pre1)
acc1
```

```python
model2=SVC(C=0.5,kernel='rbf',gamma=10.5)
```

```python
model2.fit(x_train1,y_train1)
```

```python
y_pre2=model2.predict(x_test1)
```

```python
acc2=accuracy_score(y_test1,y_pre2)
acc2
```
