# Annotating Images

## In This notebook we will cover how to annotate images using OpenCV. We will learn how to perform the following annotating

<h3> 1.Draw Lines</h3>
<h3> 2.Draw Circles</h3>
<h3> 3.Draw Rectangles</h3>
<h3> 4.Add text</h3>



```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from  IPython.display import Image
```

```python
img_python=cv.imread('python.jpg',cv.IMREAD_COLOR) # It Read in BGR

plt.imshow(img_python[:,:,::-1]) # It Prints in RGB
```

## Drawing A Line

```python
img_line=img_python[:,:,::-1].copy() # It reads in rgb format

#The Line starts from (1000,1500) and ends at (300,1500)
#The Color Of Line is RED 
#Thickness of Line IS 50px
# Line Type is cv.LINE_AA

cv.line(img_line,(1500,1500),(3500,1500),(255,0,0),thickness=50,lineType=cv.LINE_AA)

plt.imshow(img_line)
```

## Drawing a Circle

### Let's Start off by drawing a circle on an image . We will use cv2.circle function for this

## Function Syntax

### img=cv.circle(img,center,radius,color,thickness,lineype)

```python
img_circle=img_python[:,:,::-1].copy()
plt.imshow(img_circle)

cv.circle(img_circle,(2500,1500),1000,(255,0,0),thickness=50,lineType=cv.LINE_AA)
plt.imshow(img_circle)
```

# Drwaing A Rectangle

```python
img_rect=img_python[:,:,::-1].copy()

cv.rectangle(img_rect,(1500,500),(3200,2500),(255,0,0),thickness=50,lineType=cv.LINE_AA)
plt.imshow(img_rect)
```

## Adding Text On Image

<h4>Function Syntax

img = cv.putText(img,text,org,fontFace,fontScale,color,thickness,linetype,bottomLeftOrigin)</h4>

```python
img_txt=img_python[:,:,::-1].copy()
text="This Image Of Hitesh Chaudhary..!!!"
fontScale=15
fontFace=cv.FONT_HERSHEY_PLAIN
fontColor=(255,0,0)
fontThickness=25

cv.putText(img_txt,text,(1000,500),fontFace,fontScale,fontColor,fontThickness,cv.LINE_AA)
plt.imshow(img_txt)
```
