# Basic Image Enhancement Using Mathematical Operations

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

## Original Image

```python
img_cola_bgr=cv.imread('coca-cola.png',cv.IMREAD_COLOR)
plt.imshow(img_cola_bgr)

img_cola_rgb=cv.cvtColor(img_cola_bgr,cv.COLOR_BGR2RGB)

plt.imshow(img_cola_rgb)
                       
```

## Addition Or Sustract Brightness

<h4>The first operation we discuss is simple addition of images.<p>This result increasing or decreasing the brightness of the image since we are evaluating increasing or decreasing the intesity value of each pixel by the same amount,So,This will result in a global increase/decrease in brightness.</h4>

```python
matrix = np.ones(img_cola_rgb.shape,dtype="uint8") * 70
```

```python
img_cola_rgb_brighter=cv.add(img_cola_rgb,matrix)
img_cola_rgb_darker=cv.subtract(img_cola_rgb,matrix)


plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(img_cola_rgb_brighter);plt.title("Brighter Image");
plt.subplot(142);plt.imshow(img_cola_rgb_darker);plt.title("Darker Image");
plt.subplot(143);plt.imshow(img_cola_rgb);plt.title("Original");
```

<h1> Multiplication Or Contrast</h1>

<h4> Just Like addition can result in brighness change,multiplication can be used to improve the contrase of the image.</h4>

```python
matrix1=np.ones(img_cola_rgb.shape) * 0.5
matrix2=np.ones(img_cola_rgb.shape) * 5

img_cola_rgb_darker_1=np.uint8(cv.multiply(np.float64(img_cola_rgb),matrix1))
img_cola_rgb_brighter_1=np.uint8(cv.multiply(np.float64(img_cola_rgb),matrix2))

plt.figure(figsize=[20,5])
plt.subplot(131);plt.imshow(img_cola_rgb_brighter_1);plt.title("higher Contrast");
plt.subplot(132);plt.imshow(img_cola_rgb_darker_1);plt.title("Lower Contrast ");
plt.subplot(133);plt.imshow(img_cola_rgb);plt.title("Original");
```

```python
img_read=cv.imread("bulding.jpeg",cv.IMREAD_GRAYSCALE)

retval,img_thresh = cv.threshold(img_read,100,255,cv.THRESH_BINARY)

plt.figure(figsize=[18,5])
plt.subplot(121);plt.imshow(img_read,cmap='gray');plt.title("Original");
plt.subplot(122);plt.imshow(img_thresh,cmap='gray');plt.title("Thresholded");
```

```python
img_read=cv.imread('piano.jpeg',cv.IMREAD_GRAYSCALE)

retval,img_thresh_gbl_1=cv.threshold(img_read,50,255,cv.THRESH_BINARY)

retvla,img_thresh_gbl_2=cv.threshold(img_read,130,255,cv.THRESH_BINARY)

img_thresh_adp=cv.adaptiveThreshold(img_read,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,11,7)


plt.figure(figsize=(18,10))

plt.subplot(241);plt.imshow(img_read,cmap='gray');plt.title("Original");
plt.subplot(242);plt.imshow(img_thresh_gbl_1,cmap='gray');plt.title("Threshholded (global:50)");
plt.subplot(243);plt.imshow(img_thresh_gbl_2,cmap='gray');plt.title("Thresholded (global:130)");
plt.subplot(244);plt.imshow(img_thresh_adp,cmap='gray');plt.title("Threshhold (adaptive)");


```
